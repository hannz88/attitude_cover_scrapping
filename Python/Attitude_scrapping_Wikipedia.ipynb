```python
# one method
test_url = "https://en.wikipedia.org/wiki/Boy_George"
r = requests.get(test_url)
soup = BeautifulSoup(r.text, 'html.parser')
# table = soup.find('table', class_='infobox biography vcard')
table = soup.find('table', class_=re.compile(r'infobox \w* vcard'))
# print(table)
result = {}
exceptional_row_count = 0
for tr in table.find_all('tr'):
    if tr.find('th') and tr.find('td'):
        result[tr.find('th').text] = tr.find('td').text
    else:
        # the first row Logos fall here
        exceptional_row_count += 1
result
```

```python
# get the birthday
for line in t2_tab:
#     print(type(line))
    for a in line.find_all('span', { "class" : "bday" }):
        print(a.text)
```

```python
# get birthplace
for line in t2_tab:
#     print(type(line))
    for a in line.find_all('div', { "class" : "birthplace" }):
        print(a.text)
```

```python
# get occupation I
for block in t2_tab:
#     print(type(line))
    for line in block.find_all('li'):
        if line.find('a'):
            continue
        print(line.text)
```

```python
# get Occupation 2
for line in t2_tab:
#     print(type(line))
    for table_heading in line.find_all('th', { "class" : "infobox-label" }):
        table_row = table_heading.parent
        table_data = table_row.find('td')
        if table_heading.text == "Occupations":
            print(", ".join((item.text for item in table_data.find('ul').find_all("li"))))
#         print(f"{table_heading.text}: {table_data}")
```

```python
# get Years Active
for line in t2_tab:
#     print(type(line))
    for table_heading in line.find_all('th', { "class" : "infobox-label" }):
        table_row = table_heading.parent
        table_data = table_row.find('td')
        if table_heading.text.encode('utf8') == b'Years\xc2\xa0active':
            print(table_data.text)
```
